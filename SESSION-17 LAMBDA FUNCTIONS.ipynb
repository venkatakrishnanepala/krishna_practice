**case-3:if else**

```python
lambda arg1: out if else out
```

```python
def even_odd(x):
    if x%2==0:
        print('even')
    else:
        print('odd')
even_odd(11)
```

```python
even_odd1=lambda x: 'even' if x%2==0 else 'odd'
even_odd(2)
```

```python
def grade(i):
    if i>=85:print('A')
    elif i>=70:print('B')
    elif i>=50:print('C')
    else:print('D')
grade(80)
```

**case-4:if elif-else**

```python
check_grade=lambda x: 'A' if x >=85 else \
                      'B' if x >=70 else \
                      'C' if x >=50 else 'D'
check_grade(70)

```
