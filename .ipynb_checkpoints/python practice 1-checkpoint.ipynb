- variables can be **upper case**
 - NUMBER=100
- variables can be **lower case**
 - number=200
- variable are **case sensitive**
 - NUMBER=100,number=200
- variable can not allow **number as prefix**
 -123number=300
- variables can allow **number as sefix**
 - number123=400
- vaiables can not allow **spl characters**
 -number&one=400
- variables can allow only **under score**
 -number_one=500
