# **case-3:for-if-else**

```python
case-1:[<out> <for loop>]
case-2:[<out> <for loop>] <if cond>]
case-3:[<if out> <if cond> else <else out> <for loop>]
```

```python
l=[1000,2000,550,1700,200]
out=[]
for i in l:
    if i>=1000:
        out.append('exp')
    else:
        out.append('affortable')
out
```

```python
l=[1000,2000,550,1700,200]
out=['exp' if i>=1000 else 'affortable' for i in l]
out
```

```python
l=[100000,150000,2,1,60000]
out=[]
for i in l:
    if i>=60000:
        out.append('highly expensive')
    else:
        out.append('easy to pay')
out 
```
