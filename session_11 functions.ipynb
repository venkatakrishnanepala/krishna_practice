**local vs global variables**

- local variables means the variables which are used inside the function

- local variables can not use outside the function

- global variables means the variables which are used outside the function

- global variables can use anywhere in entire python script

- which means global variables can use by multiple function inside

- global variables can also use outside the functio as usual

```python
def addition():
    a=10
    b=20
    s=a+b
    print(s)
addition()
# local : a,b,s
```

```python
A=10
B=20
def addition():
    print(A)
    print(B)
    S=A+B
    print(S)
addition()
# local : s
# global : A,B
```
